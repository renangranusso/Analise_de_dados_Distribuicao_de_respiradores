```python
import pandas as pd
import seaborn as sb
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

filename = 'distribuicao_respiradores.csv' #arquivo na pasta usu√°rios
df = pd.read_csv(filename, sep=";") #sep == separador para colunas
```

```python
df.info()
```

```python
Graf = df.groupby('FORNECEDOR').size().sort_values().tail(10).plot(title='N√∫mero total de Respiradores distribuidos por Fornecedores', kind='barh', figsize=(10,5))
```

```python
graf = df.groupby('DESTINO').size().sort_values()

n = graf.count()

graf.tail(int(n)).plot(title="Respiradores recebidos por Destino", kind='barh', figsize=(10,10))
```

```python
for index, row in df.iterrows():
    if row['DESTINO'] != 'S√ÉO PAULO':
        df.drop(index, inplace=True)  #Removendo tudo o que n√£o √© Sao Paulo
        
graf = df.groupby('FORNECEDOR').size().sort_values()

n = graf.count()
        
graf.tail(int(n)).plot(title="Numero de respiradores entregues por cada fornecedor em S√£o Paulo", kind='barh', figsize=(10,5))
```
